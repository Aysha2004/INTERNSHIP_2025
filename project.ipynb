```python
insurance_dataset.replace({'sex':{'male':0 , 'female' :1}}, inplace = True)
insurance_dataset.replace({'smoker':{'yes':0 , 'no' :1}}, inplace = True)
insurance_dataset.replace({'region':{'southeast':0 , 'southwest':1 , 'northeast':2 , 'northwest':3}}, inplace = True)
```

```python
insurance_dataset.head()
```

```python
x = insurance_dataset.drop(columns='charges',axis=1)
y = insurance_dataset['charges']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
print('Accuracy Of Linear Regression Classifier on Training set:{:.2f}' .format(model.score(x_train,y_train)))
```

```python
print('Accuracy Of Linear Regression Classifier on Training set:{:.2f}' .format(model.score(x_test,y_test)))
```

```python
training_data_prediction = model.predict(x_train)
```

```python
r2_train = metrics.r2_score(y_train,training_data_prediction)
print('Required Value',r2_train)
```

```python
testing_data_prediction = model.predict(x_test)
```

```python
r2_test = metrics.r2_score(y_test,testing_data_prediction)
print('Required Value',r2_test)
```
