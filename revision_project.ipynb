```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('experience_salary.csv')
data
```

```python
#visualizing with line plot
plt.plot(data['YearsExperience'], data['Salary'], marker = 'o', linestyle = '--', color = 'blue')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('Experience vs Salary')
plt.show()
```

```python
#LinearRegression and fit
x = data[['YearsExperience']]
y = data['Salary']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)

model = LinearRegression()
model.fit(x_train, y_train)

print("coeff: {:.2f}".format(model.coef_[0]))
print("intercept: {:.2f}".format(model.intercept_))
```

```python
y_pred = model.predict(x_test)
print("Predicted salaries:", y_pred)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error

print("R² Score:", r2_score(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))
```
