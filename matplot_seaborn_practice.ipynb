```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('StudentsPerformance.csv')
data
```

```python
#scatter plot
plt.scatter(data['math score'], data['reading score'], color = 'red')
plt.xlabel('Math')
plt.ylabel('Reading')
plt.title('Math Score vs Reading Score')
plt.grid(True)
plt.show()
```

```python
#Histogram
plt.hist(data['math score'], bins = 10, color = 'green', edgecolor = 'black')
plt.xlabel('Math')
plt.ylabel('Frequency')
plt.title('Math Score Histogram')
plt.grid(True)
plt.show()
```

```python
#lmplot
sns.lmplot(x = 'math score', y = 'reading score', data = data)
plt.xlabel('Math')
plt.ylabel('Reading')
plt.title('Math Score vs Reading Score')
plt.show()
```

```python
#boxplot
sns.boxplot(x = 'gender', y = 'math score', data = data)
plt.xlabel('Gender')
plt.ylabel('Math')
plt.title('Math Score by Gender')
plt.show()
```

```python
#sns version of histogram
sns.histplot(data=data, x='math score', bins=10, kde=True, color='skyblue')
plt.title('Score Distribution')
plt.xlabel('Math')
plt.ylabel('Frequency')
plt.show()
```
